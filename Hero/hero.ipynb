```python
import pandas as pd
import numpy as np
```

```python
hero = pd.read_csv("./nuevo.csv")
```

```python
icd_codes = ["r07.0", "s93", "r05", "s01.0", "j11.1", "s83.2", "s01.0", "s92.0", "s63.", "h92.0"]
```

```python
for index, row in hero.iterrows():
    idc10 = str(row['IDC10']).lower()
    other10 = str(row['Other ICD10 code']).lower()

    if (idc10 == "other" or idc10 == "otro") and (other10 == "other" or other10 == "otro"):
        hero.at[index, "ICD 10"] = "Other"
    elif (idc10 == "other" or idc10 == "otro") and (other10 != "other" or other10 != "otro"):
        hero.at[index, "ICD 10"] = "Other"
    elif (idc10 != "other" or idc10 != "otro") and (other10 == "other" or other10 == "otro"):
        hero.at[index, "ICD 10"] = "Other"
    elif idc10 == "no hay" and other10 == "no hay":
        hero.at[index, "ICD 10"] = "¡ERROR!"
    else:
        for code in icd_codes:
            if code in idc10 and (other10 == "no hay" or other10 == "other" or other10 == "otro"):
                hero.at[index, "ICD 10"] = code
                break
            elif code in other10 and (idc10 == "no hay" or idc10 == "other" or idc10 == "otro"):
                hero.at[index, "ICD 10"] = code
                break
        else:
            hero.at[index, "ICD 10"] = "¡ERROR!"
```
